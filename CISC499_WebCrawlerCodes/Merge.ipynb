```python
combined_csv = pd.concat([pd.read_csv(f) for f in all_filenames ])# need to upload all Udemy csv files to colab before this code.
#export to csv
combined_csv.to_csv( "Udemycombined_csv.csv", index=False, encoding='utf-8-sig')
```

```python
combined_csv['domain']='Udemy'
```

```python
combined_csv.head()
```

```python
slist=[]
for column in combined_csv:
    if column=='num_reviews':
      for s in combined_csv[column]:
        string = s.replace("(","").replace(")","").replace(",","")
        slist.append(int(string))
```

```python
combined_csv['num_reviews']=slist
```
